# 1. An√°lise do faturamento


```python
# Faturamento pra cada loja
faturamento_loja1 = loja1['Pre√ßo'].sum()
faturamento_loja2 = loja2['Pre√ßo'].sum()
faturamento_loja3 = loja3['Pre√ßo'].sum()
faturamento_loja4 = loja4['Pre√ßo'].sum()
# Exibindo pra cada loja
print(f"Total loja 1: ",faturamento_loja1)
print(f"Total loja 2: ",faturamento_loja2)
print(f"Total loja 3: ",faturamento_loja3)
print(f"Total loja 4: ",faturamento_loja4)
```

# 2. Vendas por Categoria

```python
# Categoria por loja
categoria_loja1 = loja1['Categoria do Produto'].value_counts()
categoria_loja2 = loja2['Categoria do Produto'].value_counts()
categoria_loja3 = loja3['Categoria do Produto'].value_counts()
categoria_loja4 = loja4['Categoria do Produto'].value_counts()
# Exibindo categoria por loja
print(
    f"Vendas por categoria Loja 1:\n{categoria_loja1}\n\n"
    f"Vendas por categoria Loja 2:\n{categoria_loja2}\n\n"
    f"Vendas por categoria Loja 3:\n{categoria_loja3}\n\n"
    f"Vendas por categoria Loja 4:\n{categoria_loja4}"
)
```

# 3. M√©dia de Avalia√ß√£o das Lojas

```python
# Loja 1
media_loja1 = loja1['Avalia√ß√£o da compra'].mean()
media_loja1_formatada = round(media_loja1, 2)
# Loja 2
media_loja2 = loja2['Avalia√ß√£o da compra'].mean()
media_loja2_formatada = round(media_loja2, 2)
# Loja 3
media_loja3 = loja3['Avalia√ß√£o da compra'].mean()
media_loja3_formatada = round(media_loja3, 2)
# Loja 4
media_loja4 = loja4['Avalia√ß√£o da compra'].mean()
media_loja4_formatada = round(media_loja4, 2)
# Exibindo pra cada loja
print(f"M√©dia de avalia√ß√£o da loja 1: {media_loja1_formatada}")
print(f"M√©dia de avalia√ß√£o da loja 2: {media_loja2_formatada}")
print(f"M√©dia de avalia√ß√£o da loja 3: {media_loja3_formatada}")
print(f"M√©dia de avalia√ß√£o da loja 4: {media_loja4_formatada}")
```

# 4. Produtos Mais e Menos Vendidos

```python

print("Como n√£o existe uma coluna de quantidade, vamos assumir que cada linha representa uma venda individual.")
print("Portanto, o total vendido de cada produto ser√° contado pelo n√∫mero de vezes que o produto aparece no dataset.\n")

# Contagem das vendas por produto da Loja 1
vendas_por_produto_loja1 = loja1['Produto'].value_counts()

# Quantidade m√°xima e m√≠nima de vendas dos produtos da Loja 1
qtd_max_loja1 = vendas_por_produto_loja1.max()
qtd_min_loja1 = vendas_por_produto_loja1.min()

# Lista de produtos mais vendidos (podem existir empates)
mais_vendidos_loja1 = vendas_por_produto_loja1[
    vendas_por_produto_loja1 == qtd_max_loja1
].index.tolist()

# Lista de produtos menos vendidos (podem existir empates)
menos_vendidos_loja1 = vendas_por_produto_loja1[
    vendas_por_produto_loja1 == qtd_min_loja1
].index.tolist()

# Exibi√ß√£o dos resultados da Loja 1
print("üèÜ Produtos mais vendidos da LOJA 1:")
for produto in mais_vendidos_loja1:
    print(f"{produto} : {qtd_max_loja1} vendas")

print("\n‚ùå Produtos menos vendidos da LOJA 1:")
for produto in menos_vendidos_loja1:
    print(f"{produto} : {qtd_min_loja1} vendas")
```
