# 4. Produtos Mais e Menos Vendidos

```python

print("Como n√£o existe uma coluna de quantidade, vamos assumir que cada linha representa uma venda individual.")
print("Portanto, o total vendido de cada produto ser√° contado pelo n√∫mero de vezes que o produto aparece no dataset.\n")

# Contagem das vendas por produto da Loja 1
vendas_por_produto_loja1 = loja1['Produto'].value_counts()

# Quantidade m√°xima e m√≠nima de vendas dos produtos da Loja 1
qtd_max_loja1 = vendas_por_produto_loja1.max()
qtd_min_loja1 = vendas_por_produto_loja1.min()

# Lista de produtos mais vendidos (podem existir empates)
mais_vendidos_loja1 = vendas_por_produto_loja1[
    vendas_por_produto_loja1 == qtd_max_loja1
].index.tolist()

# Lista de produtos menos vendidos (podem existir empates)
menos_vendidos_loja1 = vendas_por_produto_loja1[
    vendas_por_produto_loja1 == qtd_min_loja1
].index.tolist()

# Exibi√ß√£o dos resultados da Loja 1
print("üèÜ Produtos mais vendidos da LOJA 1:")
for produto in mais_vendidos_loja1:
    print(f"{produto} : {qtd_max_loja1} vendas")

print("\n‚ùå Produtos menos vendidos da LOJA 1:")
for produto in menos_vendidos_loja1:
    print(f"{produto} : {qtd_min_loja1} vendas")
```

# Pain√©is visuais

## 1. An√°lise do faturamento (com m√©dia geral)

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Faturamento
faturamento = {
    'Loja 1': loja1['Pre√ßo'].sum(),
    'Loja 2': loja2['Pre√ßo'].sum(),
    'Loja 3': loja3['Pre√ßo'].sum(),
    'Loja 4': loja4['Pre√ßo'].sum(),
}

# M√©dia do faturamento
media_faturamento = sum(faturamento.values()) / len(faturamento)

# Cor verde
verde_top = "#c7e9b4"

plt.figure(figsize=(9,6))

# Barras do faturamento
plt.bar(faturamento.keys(), faturamento.values(), color=verde_top)

# Linha da m√©dia (tracejada)
plt.axhline(
    media_faturamento,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f"M√©dia do Faturamento: R$ {media_faturamento:,.0f}".replace(",", ".")
)

plt.title("Faturamento Total por Loja com Linha de M√©dia (Benchmark)", fontsize=16)
plt.xlabel("Lojas", fontsize=13)
plt.ylabel("Faturamento (R$)", fontsize=13)

plt.legend()

# Formatador de valores no eixo Y
plt.gca().yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, p: f"{x:,.0f}".replace(",", "."))
)

plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

## 2. Vendas por Categoria

```python
import matplotlib.pyplot as plt
import pandas as pd

# Contagem de vendas por categoria para cada loja
cat1 = loja1['Categoria do Produto'].value_counts()
cat2 = loja2['Categoria do Produto'].value_counts()
cat3 = loja3['Categoria do Produto'].value_counts()
cat4 = loja4['Categoria do Produto'].value_counts()

# Unir os dados em um √∫nico DataFrame
df_categorias = pd.DataFrame({
    'Loja 1': cat1,
    'Loja 2': cat2,
    'Loja 3': cat3,
    'Loja 4': cat4,
}).fillna(0)

# Paleta de tons de verde esmeralda
cores_verdes = [
    "#0f5132",  # verde esmeralda escuro
    "#198754",  # verde padr√£o
    "#51cf66",  # verde mais claro
    "#74d99f"   # verde bem claro
]

plt.figure(figsize=(12,6))

df_categorias.plot(
    kind='barh',
    figsize=(12,6),
    color=cores_verdes
)

plt.title("Vendas por Categoria ‚Äî Compara√ß√£o Entre Lojas")
plt.ylabel("Categoria")
plt.xlabel("Quantidade de Vendas")
plt.xticks(rotation=0)
plt.legend(title="Lojas")
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()


```

## 3. M√©dia de Avalia√ß√£o das Lojas

```python
import matplotlib.pyplot as plt

# M√©dias de avalia√ß√£o (j√° calculadas antes)
medias = {
    'Loja 1': media_loja1,
    'Loja 2': media_loja2,
    'Loja 3': media_loja3,
    'Loja 4': media_loja4
}

# Labels e valores
labels = list(medias.keys())
values = list(medias.values())

# Paleta de cor
cores = ["#c7e9b4", "#7fcdbb", "#41b6c4", "#1d91c0"]

# Gr√°fico de rosca
fig, ax = plt.subplots(figsize=(7,7))
wedges, texts, autotexts = ax.pie(
    values,
    labels=labels,
    autopct=lambda p: f'{p:.2f}%',
    startangle=90,
    wedgeprops=dict(width=0.35),
    colors=cores
)

plt.title("M√©dia de Avalia√ß√£o das Lojas", fontsize=14)
plt.show()

```

## Frete m√©dio

```python
import matplotlib.pyplot as plt

# Valores j√° calculados
fretes = {
    'Loja 1': frete_medio_loja1_formatado,
    'Loja 2': frete_medio_loja2_formatado,
    'Loja 3': frete_medio_loja3_formatado,
    'Loja 4': frete_medio_loja4_formatado
}

labels = list(fretes.keys())
values = list(fretes.values())

# Paleta de tons de verde esmeralda
cores_verdes = [
    "#0f5132",
    "#198754",
    "#51cf66",
    "#74d99f"
]

# Pequeno destaque nas fatias (explodir leve)
explode = [0.03, 0.03, 0.03, 0.03]

plt.figure(figsize=(7,7))
plt.pie(
    values,
    labels=labels,
    autopct='%.2f',
    startangle=90,
    colors=cores_verdes,
    pctdistance=0.8,              # aproxima o texto
    explode=explode,               # leve destaque
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}  # bordas modernas
)

plt.title("Frete M√©dio por Loja", fontsize=14, pad=15)
plt.tight_layout()
plt.show()

```
